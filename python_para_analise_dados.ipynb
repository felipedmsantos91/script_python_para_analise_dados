# Python para análise de dados

Análise de dados de casas no EUA.

```python
#Importando a biblioteca pandas
import pandas as pd
```

```python
#Lendo a base de dados
df = pd.read_csv('kc_house_data.csv',sep=',')
df.head()
```

```python
#Contando a quantidade de linhas do dataframe
df.count()
```

```python
#Imprimindo as informações do Dataframe
df.info()
```

```python
#Imprimindo o nome das colunas do Dataframe
df.columns
```

```python
#Imprimindo as informações estatisticas da coluna bedrooms
df['bedrooms'].describe()
```

```python
#Imprimindo os valores únicos (imprimir os valores uma unica vez) da coluna bedrooms
df['bedrooms'].unique()
```

```python
#Imprimindo o valor médio da coluna bathrooms
df['bathrooms'].mean()
```

```python
#Somando a quantidade de valores nulos de cada coluna do Dataframe
df.isnull().sum()
```

```python
#Remover as linhas onde tenha um registro faltante em algum atributo
df.dropna(subset=['floors'],inplace=True)
```

```python
df.isnull().sum()
```

```python
#Preencher com 1 os values null da coluna bedrooms
df['bedrooms'].fillna(1.0,inplace=True)
```

```python
df.isnull().sum()
```

```python
#Retornar todos os imoveis que tenham o número de bathrooms maior que 1
df[df['bathrooms']>1]
```

```python
df.loc[df['bathrooms']>1]#usando a função loc
```

```python
#Retornar todos os imoveis que tenham a quantidade de bedrooms igual a 4
df.loc[df['bedrooms']==4]
```

```python
#Contar o retorno da consulta anterior
df.loc[df['bedrooms']==4].count()
```

```python
#Ordenar o Dataframe pela coluna price em ordem decrescente
df.sort_values(by='price',ascending=False)
```

```python
# Agrupar e contar quantidade de imóveis por quantidade de quartos
df['bedrooms'].value_counts()
```

```python
#Criar uma coluna que seja a soma das colunas bathrooms e bedrooms com o nome comodos
df['sum_bath_bed'] = df['bathrooms'] + df['bedrooms']
df['sum_bath_bed']
```

```python
df.head()
```

```python
#Visualizar em formato de histograma a distribuição dos imoveis por preço
%matplotlib notebook
df.hist(column='price',bins=50)
```

```python
#Visualizar em formato de disperção os imoveis utilizando as colunas bathrooms e price
df.plot(x='bathrooms',y='price',kind='scatter',grid=True)
```
